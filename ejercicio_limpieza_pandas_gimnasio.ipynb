## 🕳️ Paso 4 — Manejar valores nulos

Tenemos valores nulos en `Date` y `Calories`. Los trataremos de forma diferente.

### Ejercicio 4.1 — Ver las filas con nulos
Muestra solo las filas que tienen algún valor nulo.

> 💡 **Pista:** `df[df.isnull().any(axis=1)]`  
> `axis=1` significa "busca en cada fila"

```python
# ✏️ Muestra las filas con nulos
df [df .isnull() .any(axis=1)]
```

### Ejercicio 4.2 — Rellenar Calories con la media
Rellena los huecos de `Calories` con la media de esa columna, redondeada a 1 decimal.

> 💡 **Pistas:**
> - Calcula primero: `df['Calories'].mean()`
> - Redondea: `round(media, 1)`
> - Rellena: `.fillna(valor)`

```python
# ✏️ Calcula la media e imprímela
media_calorias = df ["Calories"] .mean()
df ["Calories"] = df ["Calories"] .fillna(round(media_calorias , 1))





print('Nulos en Calories:', df['Calories'].isnull().sum())
```

### Ejercicio 4.3 — Eliminar la fila con Date nula
La fila que no tiene fecha no nos sirve. Elimínala y reinicia el índice.

> 💡 **Pistas:**
> - `.dropna(subset=['Date'])` elimina filas con nulo en Date
> - `.reset_index(drop=True)`

```python
# ✏️ Elimina la fila sin fecha
df = df .dropna(subset=["Date"])
df = df .reset_index(drop=True)


print('Filas restantes:', len(df))
print('Nulos totales:', df.isnull().sum().sum())
```

## 📅 Paso 5 — Limpiar la columna de fechas

La columna `Date` tiene un problema: la mayoría de fechas tienen comillas (`'2020/12/01'`) y **una** tiene un formato completamente diferente (`20201226`).

### Ejercicio 5.1 — Ver los valores únicos de Date
Imprime todos los valores de la columna `Date` para ver el problema.

> 💡 **Pista:** `df['Date'].tolist()`

```python
import pandas as pd
# 1 volvemos a cargar los datos
df = pd.read_csv("dataset_gimnasio.csv")
# 2 Quitamos duplicados
df = df.drop_duplicates().reset_index(drop=True)
# 3 Arreglamos las Calorias (rellenar nulos con la media)
df["Calories"] = df["Calories"].fillna(df["Calories"].mean()).round(2)
# 4 Arreglamos las fechas (convertir a formato real y quitar nulos)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"]).reset_index(drop=True)
# 5 Corregimos el error de las 7 hors (el 450)  
df . loc[df["Duration"] == 450, "Duration"] = 45  

# --- COMPROBACIÓN FINAL ---
print("✅ ¡Tabla limpia!")
print(f"Filas finales: {len(df)}")
print(df.head(10)) # Vemos las primeras 10 filas
df.to_csv("datos_limpios.csv", index=False)
```

```python
# ✏️ Muestra el DataFrame final
# --- LIMPIEZA FINAL ---
# 1. Borrar duplicados
df.drop_duplicates(inplace=True)

# 2. Borrar la fila con fecha NaT (la 22)
df.dropna(subset=['Date'], inplace=True)

# 3. Rellenar calorías vacías con la media
df.fillna({'Calories': df['Calories'].mean()}, inplace=True)
# Redondear las calorías a 2 decimales para que quede más bonito
df['Calories'] = df['Calories'].round(2)


# 4. Resetear el índice por última vez para que sea perfecto
df.reset_index(drop=True, inplace=True)

# Mostrar el resultado final
df
```
